```python
from midiutil import MIDIFile

themeDir = "/Users/paul/Music/Logic/DisquietJunto/DQ422"
t1 = themeDir + "/Theme1.mid"
t2 = themeDir + "/Theme2.mid"
```

```python
from mxm.midifile import MidiInFile, MidiEvents
class MidiCollect(MidiEvents):
    def __init__(self):
        super().__init__()
        self.collect = []
    def note_on(self, channel=0, note=0x40, velocity=0x40, use_running_status=False):
        self.collect.append( ( "ON", self.abs_time(), note, velocity ) )
    def note_off(self, channel=0, note=0x40, velocity=0x40, use_running_status=False):
        self.collect.append( ( "OFF", self.abs_time(), note, velocity ) )
    
e = MidiCollect()
midiIn = MidiInFile(e, t1)
midiIn.read()
t1notes = e.collect

e = MidiCollect()
midiIn = MidiInFile(e, t2)
midiIn.read()
t2notes = e.collect

[t1notes, t2notes]
```

```python
# Turn that into note duration sets
def onoffToDur(nts):
    starts = {}
    durs = []

    t0 = nts[0][1]
    for o,t,n,v in nts:
        if( v == 0 ):
            (st,sv) = starts[n]
            durs.append(( st-t0, n, sv, t-st))
        else:
            starts[n] = (t,v)
    durs.sort( key = lambda x : x[0] )
    return durs
    
t1durs = onoffToDur(t1notes)
t2durs = onoffToDur(t2notes)

[t1durs, t2durs]
```

```python
# OK this actually sounds pretty good TBH. So lets pause for a bit and think about what music would be from here.
test1 = MIDIFile(1)

test1.addTempo(1000, 0, 240)
import random
random.seed(1832)

n = [t1durs, t2durs]
idx = 0
for i in range(40):
    off = i * 2
    div = 10 * ( 240 + random.randrange(480 + 240))
    nrand = random.randrange(-12, 12)
    for e in n[idx]:
        # print(0,0,e[1],(e[0]/div) + off, (e[3]/div), e[2])
        test1.addNote(0,0,e[1] + nrand,(e[0]/div) + off, (e[3]/div), e[2])
    idx = idx + 1
    if( idx > 1 ):
        idx = 0
        
with open(themeDir + "/testnew.mid", 'wb') as output_file:
    test1.writeFile(output_file)
```
